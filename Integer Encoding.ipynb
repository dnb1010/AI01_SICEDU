```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
text = "A barber is a person. a barber is a good person. a barber is a huge person. he Knew A Secret! The Secret He Kept is huge secret.Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

```python
text = sent_tokenize(text)
print(text)
```

```python
# Cleaning and word tokenization
# import nltk
# nltk.download('stopwords')
# nltk.download('punkt_tab')
vocab = {} # Python dictionary data type
sentences = []
stop_words = set(stopwords.words('english'))
for i in text:
  sentence = word_tokenize(i) # Perform word tokenization
  result = []

  for word in sentence:
    word = word.lower() # Reduce the number of words by lowering all words
    if word not in stop_words: # Eliminate stopwords from the result of word tokenization
      if(len(word) > 2): # Additionally remove the words with word length less than 2
        result.append(word)
        if word not in vocab:
          vocab[word] = 0
        vocab[word] += 1
  sentences.append(result)
print(sentences)
```

```python
print(vocab)
```

```python
print(vocab["barber"])
```

```python
vocab_sorted = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
vocab_sorted
```

```python
word_to_index = {}
i = 0
for (word, frequency) in vocab_sorted:
  if(frequency > 1): # Remove index information about the relevant words
    i += 1
    word_to_index[word] = i
word_to_index
```

```python
vocab_size = 5
words_frequency = [w for w, c in word_to_index.items() if c >= vocab_size + 1] # Remove words with index greater than 5
for w in words_frequency:
  del word_to_index[w] # Remove index information about the relevant words
word_to_index
```

```python
word_to_index['OOV'] = len(word_to_index) + 1
```

```python
word_to_index['OOV']
```

```python
len(word_to_index)
```

```python
encoded = []
for s in sentences:
  temp = []
  for w in s:
    try:
      temp.append(word_to_index[w])
    except KeyError:
      temp.append(word_to_index['OOV'])
  encoded.append(temp)
  print(encoded)
```

```python
print(sentences)
```

```python
words = sum(sentences, [])
# It is possible to use words = np.hstack(sentences)
print(words)
```

```python
vocab = Counter(words) # Easily calculate the frquency of words by using Python Counter module
print(vocab)
```

```python
print(vocab["barber"])
```

```python
vocab_size  = 5
vocab = vocab.most_common(vocab_size) # Save only the top five frequency words
vocab
```

```python
word_to_index = {}
i = 0
for (word, frequency) in vocab:
  i += 1
  word_to_index[word] = i
print(word_to_index)
```

**Use Keras**

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentences=[['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
# fit_on_texts(), generate the group of words based on frequency
```

```python
print(tokenizer.word_index)
```

```python
print(tokenizer.word_counts)
```

```python
print(tokenizer.texts_to_sequences(sentences))
```
